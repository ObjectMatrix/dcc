📋 PART 1 — LISTS: ORDERED AND CHANGEABLE
----------------------------------------

A **list** is like a numbered shelf. Items have positions (indexes), and you can add, remove, or change items any time.

```
planets = ["Mercury", "Venus", "Earth", "Mars"]
index 0     index 1   index 2   index 3
```

⚠️ **Index starts at 0, not 1.** This trips everyone up at first.

````python
# 🌍 Creative Example: Mars Mission Crew Roster
# The crew list changes as astronauts join and leave the mission.

crew = ["Amara", "Dev", "Soo-Jin", "Carlos"]

print("=== MISSION CREW ===")
print("Full crew:", crew)
print("Crew size:", len(crew))
print("Commander (first):", crew[0])
print("Engineer (last):", crew[-1])     # -1 always means the last item
````

````python
# Lists are MUTABLE — you can change them

crew = ["Amara", "Dev", "Soo-Jin", "Carlos"]

# Add a new crew member
crew.append("Yuki")
print("After adding Yuki:", crew)

# Remove a crew member
crew.remove("Dev")
print("After Dev left:", crew)

# Replace someone
crew[0] = "Fatima"   # index 0 = first item
print("New commander:", crew)

print(f"\nFinal crew of {len(crew)}: {crew}")
````

````python
# Iterating (looping) through a list

crew = ["Fatima", "Soo-Jin", "Carlos", "Yuki"]

print("=== CREW BRIEFING ===")
for member in crew:
    print(f"  Welcome aboard, {member}! 🚀")

print(f"\nAll {len(crew)} crew members checked in.")
````

🔒 PART 2 — TUPLES: ORDERED BUT LOCKED
-------------------------------------

A **tuple** is like a list, but **immutable** — once created, it cannot be changed.

Use tuples for things that *should not* change: coordinates, dates, fixed records.

```python
list  = [1, 2, 3]   # square brackets — changeable
tuple = (1, 2, 3)   # round brackets — locked
```

````python
# 🌍 Creative Example: Planet Data Records
# Planet facts don't change — perfect for tuples!

# (name, moons, diameter_km, has_rings)
earth   = ("Earth",   1,   12_742, False)
saturn  = ("Saturn",  83,  116_460, True)
jupiter = ("Jupiter", 95,  139_820, True)

print("=== PLANET RECORDS ===")
for planet in [earth, saturn, jupiter]:
    name, moons, diameter, rings = planet    # unpacking!
    ring_text = "has rings" if rings else "no rings"
    print(f"{name}: {moons} moons, {diameter:,} km diameter, {ring_text}")
````

````python
# Tuples are IMMUTABLE — this will cause an error (intentionally!)
# Read the error message carefully.

launch_coordinates = (28.5721, -80.6480)   # Kennedy Space Center GPS
print("Launch site coordinates:", launch_coordinates)
print("Latitude:", launch_coordinates[0])
print("Longitude:", launch_coordinates[1])

# Uncomment the line below to see the error:
# launch_coordinates[0] = 99.999   # TypeError: tuples don't allow this!
````

🗂️ PART 3 — DICTIONARIES: DATA WITH LABELS
------------------------------------------

A **dictionary** stores data as **key → value** pairs. Instead of remembering index numbers, you look up data by name.

```python
astronaut = {
"name": "Amara",
"age": 14,
"role": "Commander"
}
```

Think of it like a real dictionary: you look up a **word** (key) to find its **definition** (value).

````python
# 🌍 Creative Example: Astronaut Profile System

astronaut = {
    "name":     "Amara Osei",
    "age":      14,
    "role":     "Commander",
    "missions": 3,
    "origin":   "Ghana"
}

print("=== ASTRONAUT PROFILE ===")
print(f"Name:     {astronaut['name']}")
print(f"Age:      {astronaut['age']}")
print(f"Role:     {astronaut['role']}")
print(f"Missions: {astronaut['missions']}")
print(f"Origin:   {astronaut['origin']}")
````

````python
# Dictionaries are mutable — you can add, change, and remove entries

astronaut = {
    "name":  "Amara Osei",
    "age":   14,
    "role":  "Commander"
}

# Add a new key
astronaut["specialty"] = "Robotics"
print("After adding specialty:", astronaut)

# Update an existing key
astronaut["missions"] = 4
print("After mission update:", astronaut)

# Safe lookup with .get() — won't crash if key doesn't exist
height = astronaut.get("height", "Not recorded")
print("Height:", height)   # returns default instead of error
````

````python
# What happens when you look up a key that doesn't exist?
# Without .get() — this causes a KeyError!

astronaut = {"name": "Amara", "age": 14}

# Safe way:
print(astronaut.get("height", "Key not found"))   # no crash

# Uncomment to see the crash:
# print(astronaut["height"])   # KeyError!
````

````python
# Iterating through a dictionary

astronaut = {
    "name":      "Amara Osei",
    "age":       14,
    "role":      "Commander",
    "specialty": "Robotics",
    "origin":    "Ghana"
}

print("=== FULL PROFILE ===")
for key, value in astronaut.items():
    print(f"  {key.capitalize()}: {value}")
````

🎯 PART 4 — SETS: UNIQUE VALUES ONLY
-----------------------------------

A **set** automatically removes duplicates. Every item appears exactly once.

Sets are also great for **comparing** two groups — what do they share? What's different?

```python
tags = {"robotics", "coding", "space", "coding"}  # duplicate removed!
RESULT: {"ROBOTICS", "CODING", "SPACE"}
---------------------------------------
```

⚠️ Sets are **unordered** — the items have no index position.

````python
# 🌍 Creative Example: Mission Skills Tracker
# Track what skills each crew member brings — no duplicates needed!

amara_skills  = {"robotics", "navigation", "python", "first-aid"}
carlos_skills = {"python", "mechanics", "first-aid", "geology"}

print("Amara's skills:", amara_skills)
print("Carlos's skills:", carlos_skills)

# What skills do BOTH have? (intersection)
shared = amara_skills & carlos_skills
print("\nShared skills:", shared)

# What skills does EITHER have? (union)
all_skills = amara_skills | carlos_skills
print("All crew skills:", all_skills)

# What does Amara have that Carlos doesn't? (difference)
unique_to_amara = amara_skills - carlos_skills
print("Only Amara has:", unique_to_amara)
````

````python
# Sets remove duplicates automatically
# Great for cleaning messy data!

sensor_readings = [22.1, 22.3, 22.1, 25.0, 22.3, 19.8, 25.0, 22.1]
print("Raw readings:", sensor_readings)
print("Count:", len(sensor_readings))

unique_readings = set(sensor_readings)
print("\nUnique readings only:", unique_readings)
print("Unique count:", len(unique_readings))
````

🧩 PART 5 — PUTTING IT ALL TOGETHER
----------------------------------

Real programs mix all four structures. Here's a mission database that uses all of them.

````python
# 🌍 Full Example: Mars Mission Database
# Lists hold crew, dicts hold profiles, tuples hold fixed data, sets hold tags

# Fixed mission info (tuple — doesn't change)
mission_info = ("Ares I", 2047, "Mars", 225_000_000)
mission_name, launch_year, destination, distance_km = mission_info

# Crew roster (list — changes as people join/leave)
crew_roster = ["Amara", "Carlos", "Yuki"]

# Crew profiles (dict — labelled data)
crew_profiles = {
    "Amara":  {"role": "Commander",  "age": 14, "origin": "Ghana"},
    "Carlos": {"role": "Engineer",   "age": 15, "origin": "Mexico"},
    "Yuki":   {"role": "Scientist",  "age": 13, "origin": "Japan"}
}

# Mission tags (set — unique only)
mission_tags = {"science", "exploration", "international", "youth", "science"}

# Print the full mission brief
print(f"{'=' * 40}")
print(f"  MISSION: {mission_name}")
print(f"  Launch: {launch_year} | Destination: {destination}")
print(f"  Distance: {distance_km:,} km")
print(f"  Tags: {mission_tags}")
print(f"{'=' * 40}")
print(f"CREW ({len(crew_roster)} members):")
for name in crew_roster:
    profile = crew_profiles[name]
    print(f"  {name} | {profile['role']} | Age {profile['age']} | {profile['origin']}")
print(f"{'=' * 40}")
````

🧠 DATA STRUCTURE CHEAT SHEET
----------------------------

| Structure | Syntax | Ordered? | Changeable? | Duplicates? | Best for |
|-----------|--------|----------|-------------|-------------|----------|
| List | `[1, 2, 3]` | ✅ Yes | ✅ Yes | ✅ Yes | Sequences that change |
| Tuple | `(1, 2, 3)` | ✅ Yes | ❌ No | ✅ Yes | Fixed records, coordinates |
| Dictionary | `{"k": v}` | ✅ Yes* | ✅ Yes | Keys: ❌ | Labelled data, profiles |
| Set | `{1, 2, 3}` | ❌ No | ✅ Yes | ❌ No | Unique values, comparisons |

### 🛠️ Mutability — How to Add, Remove, Update

| Type | Mutability | Add | Remove | Update |
|------|------------|-----|--------|--------|
| **List** | ✅ Mutable | `lst.append(x)` · `lst.insert(i, x)` | `lst.remove(x)` · `lst.pop(i)` · `del lst[i]` | `lst[i] = x` |
| **Tuple** | ❌ Immutable | ❌ Not allowed (rebuild: `t + (x,)`) | ❌ Not allowed (rebuild: slice) | ❌ Not allowed (rebuild a new tuple) |
| **Dictionary** | ✅ Mutable | `d["k"] = v` · `d.update({...})` | `del d["k"]` · `d.pop("k")` | `d["k"] = new_v` |
| **Set** | ✅ Mutable | `s.add(x)` · `s.update({...})` | `s.remove(x)` · `s.discard(x)` | ❌ No update — remove + add |

---

🔑 KEY METHODS TO REMEMBER
-------------------------

```python
LIST
----
my_list.append(item)      # add to end
my_list.remove(item)      # remove by value
my_list[0]                # access by index

DICT
----
my_dict["key"]            # access by key (crashes if missing)
my_dict.get("key", "?")   # safe access with default
my_dict.items()           # loop key + value pairs

SET
---
set_a & set_b             # intersection (shared)
set_a | set_b             # union (all)
set_a - set_b             # difference (only in a)

ALL
---
len(collection)           # number of items
for item in collection    # loop through any of them
```
