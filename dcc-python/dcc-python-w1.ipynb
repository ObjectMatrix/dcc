```

age = 13 ← the box is called 'age', it holds the number 13
```

## 🔢 Part 2 — Data Types

Python has 4 basic data types you'll use constantly:

| Type | Example | What it is |
|------|---------|------------|
| `int` | `13` | Whole number |
| `float` | `3.14` | Decimal number |
| `str` | `"hello"` | Text (string) |
| `bool` | `True` | Yes/No value |

Use `type()` to ask Python what type something is.

````python
# Python can tell you the type of any value

print(type(42)) # int
print(type(3.14)) # float
print(type("hello")) # str
print(type(True)) # bool
````

````python
# ⚠️ The classic trap: input() always returns a STRING
# Even if someone types 25, Python sees "25" not 25
# You must CONVERT it!

age_text = "17" # this looks like a number but it's a string
age_number = int(age_text) # int() converts string to integer

print("Text version:", age_text, "— type:", type(age_text))
print("Number version:", age_number, "— type:", type(age_number))

# Now we can do math with it
years_to_mars_mission = 2047 - age_number
print("Years until your Mars mission:", years_to_mars_mission)
````

## ➕ Part 3 — Arithmetic Operators

| Operator | Symbol | Example | Result |
|----------|--------|---------|--------|
| Add | `+` | `5 + 3` | `8` |
| Subtract | `-` | `10 - 4` | `6` |
| Multiply | `*` | `3 * 4` | `12` |
| Divide | `/` | `10 / 3` | `3.333` |
| Floor divide | `//` | `10 // 3` | `3` |
| Modulo | `%` | `10 % 3` | `1` |
| Power | `**` | `2 ** 8` | `256` |

The **modulo** `%` is a secret weapon — it gives you the **remainder** after division. We already used it for leap years!

````python
# 🌍 Creative Example: Rocket Fuel Calculator
# The Tsiolkovsky rocket equation tells us how much fuel we need.
# Here's a simplified version for our mission.

distance_to_mars_km = 225_000_000 # average distance in km (underscore = readable large numbers)
speed_kmh = 28_000 # spacecraft speed in km/h

travel_hours = distance_to_mars_km / speed_kmh
travel_days = travel_hours / 24
travel_months = travel_days / 30

print("=== MARS TRAVEL CALCULATOR ===")
print(f"Distance: {distance_to_mars_km:,} km")
print(f"Speed: {speed_kmh:,} km/h")
print(f"Travel time: {travel_hours:,.0f} hours")
print(f"That's about {travel_days:.0f} days")
print(f"Or roughly {travel_months:.1f} months!")
````

````python
# The modulo operator % — checking divisibility
# If number % divisor == 0, it divides evenly (no remainder)

year = 2024

if year % 100 == 0: # century year?
    if year % 400 == 0:
        print(year, "is a leap year (century rule passed)")
    else:
        print(year, "is NOT a leap year (century exception)")
else:
    if year % 4 == 0:
        print(year, "is a leap year")
    else:
        print(year, "is NOT a leap year")
````

## 💬 Part 4 — Strings: Working with Text

Strings are text. You can join them, slice them, measure them, and format them.

The most powerful way to format text is an **f-string** — put `f` before the quotes and use `{}` to insert variables directly.

````python
# String basics

mission = "Apollo"
number = 11

# Old way (concatenation) — clunky
print("Mission: " + mission + " " + str(number))

# f-string way — clean and powerful
print(f"Mission: {mission} {number}")
print(f"Mission name has {len(mission)} letters")
print(f"Uppercase: {mission.upper()}")
print(f"Lowercase: {mission.lower()}")
````

````python
# f-strings can do math inside the curly braces too!

price_per_kg = 54_500 # cost to send 1kg to space (USD)
luggage_kg = 23

print(f"Your {luggage_kg}kg suitcase would cost ${price_per_kg * luggage_kg:,} to send to space.")
print(f"Maybe pack lighter. 🧳")
````
